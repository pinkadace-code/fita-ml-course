**Problēmas apraksts.**
Projektā tiek izmantoti dati no interneta veikala. Datu kopā ir 12 330 apmeklējumu ieraksti ar informāciju par to, kādas lapas apmeklētājs skatījās, cik ilgi uzturējās vietnē un vai apmeklējums notika tuvu svētkiem.
Mērķis ir prognozēt, vai konkrētais apmeklētājs veiks vai neveiks pirkumu.
Šāda prognoze ir vērtīga uzņēmumam, jo ļauj savlaicīgi atpazīt potenciālos pircējus un piedāvāt viņiem personalizētas atlaides, vai citus izdevīgus piedāvājumus, tādējādi palielinot pārdošanas apjomu.

**ML tipa identificēšana.** Vai ir "pareizā atbilde" datos? Jā — kolonna Revenue norāda, vai ir veikts pirkums (True/False) katra konkrētā pircēja apmeklējumā.
Kad mēs gribam atbildi? Katram jaunam apmeklētājam individuāli.
ML veids: Supervised learning.
Apakštips: Classification (atbilde ir "nopirka" vai "nenopirka")

**Target un features.** Target mainīgais: Revenue — True/False vērtība — vai apmeklētājs nopirka/nenopirka.
Features: skaitliskas - PageValues (cik vērtīga bija lapa ceļā uz pirkumu), BounceRates (cik apmeklētāji aizgāja jau pēc pirmās lapas apskates), ExitRates (cik apmeklētāji aizgāja no konkrētās lapas), ProductRelated_Duration (cik ilgi apmeklētājs skatījās preces); kategoriskas - Month, VisitorType, Weekend

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
print("✓ Visas bibliotēkas ielādētas veiksmīgi")
```

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00468/online_shoppers_intention.csv"
df = pd.read_csv(url)

print(f"Datu kopu forma: {df.shape}")
print(f"Rindu skaits: {df.shape[0]}")
print(f"Kolonnu skaits: {df.shape[1]}")
print("\nPirmās 5 rindas:")
print(df.head())
```

```python
# Datu tipi
print("Datu tipi:")
print(df.dtypes)
```

```python
# Statistika par skaitliskajām kolonnām
print("Skaitlisko kolonnu statistika:")
print(df.describe())
```

```python
# Trūkstošas vērtības
print("Trūkstošas vērtības:")
print(df.isnull().sum())
```

```python
cat_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()
print(f"Kategoriskās kolonnas: {cat_cols}")
```

```python
print(df['Revenue'].value_counts())
print(f"Pirkumu īpatsvars: {df['Revenue'].sum() / len(df) * 100:.2f}%")
```

```python
# Kuras kolonnas satur trūkstošas vērtības?
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

print("Kolonnas ar trūkstošām vērtībām:")
print(missing)

# Vizualizācija
if len(missing) > 0:
    plt.figure(figsize=(10, 4))
    missing.plot(kind='barh', color='coral')
    plt.title('Trūkstošās vērtības pa kolonnām')
    plt.xlabel('Trūkstošo vērtību skaits')
    plt.tight_layout()
    plt.show()
else:
    print("✓ Trūkstošo vērtību nav — dati ir tīri!")
```

```python
# Skaitliskām kolonnām — mediāna
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"✓ {col}: aizpildīts ar mediānu {median_val:.2f}")

# Kategoriskām kolonnām — moda
cat_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col] = df[col].fillna(mode_val)
        print(f"✓ {col}: aizpildīts ar modu '{mode_val}'")

# Galīgā pārbaude
assert df.isnull().sum().sum() == 0, "Vēl ir trūkstošas vērtības!"
print("\n✓ Visas trūkstošas vērtības apstrādātas. df.isnull().sum().sum() =", df.isnull().sum().sum())
```

Skaitliskajās kolonnās trūkstošās vērtības tiek aizstātas ar mediānu, jo ekstrēmas vērtības to neietekmē.
Kategoriskajās kolonnās trūkstošās vērtības tiek aizstātas ar modu, jo mediānu nevar aprēķināt - ja nezinām apmeklētāja tipu, pieņemam, ka tas ir visizplatītākais.

```python
# Kādas ir kategoriskās kolonnas?
print("Kategoriskās kolonnas:")
cat_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()
print(cat_cols)

for col in cat_cols:
    print(f"\n{col}: {df[col].nunique()} unikālas vērtības")
    print(df[col].unique()[:10])  # Parāda pirmos 10
```

```python
# Boolean/binary kolonnas — Label Encoding
df['Weekend'] = df['Weekend'].astype(int)
df['Revenue'] = df['Revenue'].astype(int)
print("✓ Boolean kolonnas konvertētas uz 0/1")
```

```python
# One-Hot Encoding
df = pd.get_dummies(df, columns=['Month', 'VisitorType'], drop_first=True)
print(f"✓ One-Hot Encoding pabeigts. Jaunas kolonnas: {df.shape[1]}")
print("Nākamās jaunās kolonnas:")
print(df.columns[df.columns.str.contains('Month|VisitorType')])
```

```python
# Pārliecinies, ka nav palikušas neapstrādātas kategoriskās kolonnas
remaining_cat = df.select_dtypes(include=['object', 'bool']).columns.tolist()
assert len(remaining_cat) == 0, f"Vēl neapstrādātas: {remaining_cat}"
print("✓ Visas kategoriskās kolonnas kodētas. Atlikt: ", len(remaining_cat))

# Visas kolonnas tagad ir skaitliskas
print(f"\nKolonnu skaits pēc kodēšanas: {df.shape[1]}")
print("Dati tipi:")
print(df.dtypes.value_counts())
```

pd.get_dummies() izveidoja jaunās kolonnas, bet tās ir bool tipa — True/False — nevis 0/1. Tāpēc assert tās vēl uzskata par neapstrādātām.
Claude piedāvā bool kolonnu labošanas kodu

```python
# Pārvēršam esošās bool kolonnas uz int
bool_cols = df.select_dtypes(include=['bool']).columns.tolist()
df[bool_cols] = df[bool_cols].astype(int)
print(f"✓ {len(bool_cols)} bool kolonnas konvertētas uz 0/1")
```

```python
remaining_cat = df.select_dtypes(include=['object', 'bool']).columns.tolist()
assert len(remaining_cat) == 0, f"Vēl neapstrādātas: {remaining_cat}"
print("✓ Visas kategoriskās kolonnas kodētas. Atlikušas: ", len(remaining_cat))
print(f"\nKolonnu skaits pēc kodēšanas: {df.shape[1]}")
print("Datu tipi:")
print(df.dtypes.value_counts())
```

Boolean kolonnām (Weekend un Revenue) izmantots Label encoding, jo šīm kolonnām ir tikai divas vērtības — True vai False.
Kolonnām ar vairākām kategoriskām vērtībām (Month un VisitorType) izmantota One-Hot Encoding, lai katra unikāla vērtība dabūtu savu kolonnu ar 0 vai 1.
Rezultāts: no 18 sākotnējām kolonnām izveidotas 27 kolonnas, visas ir skaitliskas.

```python
# Kādas ir skaitliskās kolonnas
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Skaitliskās kolonnas ({len(numeric_cols)} pavisam):")
print(numeric_cols[:10])  # Parāda pirmos 10

# Statistika PIRMS normalizācijas
print("\n=== STATISTIKA PIRMS NORMALIZĀCIJAS ===")
print(df[numeric_cols[:6]].describe().round(2))
```

```python
from sklearn.preprocessing import StandardScaler

# Atlasi skaļošanai VISAS skaitliskās kolonnas (izņemot target)
cols_to_scale = [col for col in numeric_cols if col != 'Revenue']

# StandardScaler: (x - mean) / std
scaler = StandardScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

print(f"✓ {len(cols_to_scale)} skaitliskās kolonnas normalizētas")
print("\n=== STATISTIKA PĒC NORMALIZĀCIJAS ===")
print(df[cols_to_scale[:6]].describe().round(2))
```

```python
# Salīdzinājums:
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()

sample_cols = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration',
               'BounceRates', 'ExitRates', 'PageValues']

for idx, col in enumerate(sample_cols):
    if col in df.columns:
        axes[idx].boxplot(df[col], vert=True)
        axes[idx].set_title(f'{col} (pēc normalizācijas)')
        axes[idx].set_ylabel('Normalizētā vērtība')

plt.tight_layout()
plt.show()

print("✓ Normalizācijas efekts redzams grafikos — vērtības ir tuvu 0 ar std ≈ 1")
```

```python
# Iesaistīšanās rādītājs — cik aktīvi apmeklētājs mijiedarbojās ar vietni
df['Engagement_Score'] = df['ProductRelated_Duration'] - df['BounceRates']

print("✓ Jaunas iezīmes izveidotas")
print(f"Datu kopas forma tagad: {df.shape}")
```

Jaunā iezīme - Engagement_Score
Engagement_Score mēra apmeklētāja iesaistīšanās līmeni vietnē — cik nopietni viņš aplūkoja preces. Augsta vērtība nozīmē, ka apmeklētājs ilgi skatījās preces un reti aizgāja no lapām, kas liecina par patiesu interesi. Zema vai negatīva vērtība nozīmē, ka apmeklētājs ātri aizgāja, neiepazīstoties ar precēm.

Jo augstāks Engagement_Score, jo lielāka varbūtība, ka apmeklētājs preci nopirks. Jaunā iezīme šo varbūtību izsaka skaitliski, līdz ar to tā var noderēt apmācot ML modeli.

```python
# Salīdzinājums: jaunā iezīme starp pircējiem un nepircējiem
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# 1. Histogramma
axes[0].hist([df[df['Revenue'] == 0]['Engagement_Score'],
              df[df['Revenue'] == 1]['Engagement_Score']],
             label=['Nav pirkuma', 'Ir pirkums'], bins=30, alpha=0.7)
axes[0].set_title('Engagement_Score: Pircēji vs. Nepircēji')
axes[0].set_xlabel('Engagement_Score (normalizēta)')
axes[0].legend()

# 2. Vidējie
axes[1].bar(['Nav pirkuma', 'Ir pirkums'],
            [df[df['Revenue'] == 0]['Engagement_Score'].mean(),
             df[df['Revenue'] == 1]['Engagement_Score'].mean()],
            color=['steelblue', 'coral'])
axes[1].set_title('Vidējais Engagement_Score')
axes[1].set_ylabel('Vērtība')

# 3. Box plot
df.boxplot(column='Engagement_Score', by='Revenue', ax=axes[2])
axes[2].set_title('Engagement_Score izkliede')
axes[2].set_xlabel('Pirkums (0=Nē, 1=Jā)')

plt.tight_layout()
plt.show()
print("✓ Jaunā iezīme analizēta un vizualizēta")
```

Datu sagatavošanas kopsavilkums

Identificētās problēmas:
Četras kolonnas nebija skaitliskā formātā - Month un VisitorType saturēja tekstu, Weekend un Revenue saturēja True/False vērtības.
Skaitlisko kolonnu vērtību skalas bija ļoti atšķirīgas — ProductRelated_Duration maksimums bija 63 973, bet Administrative maksimums tikai 27.
Datos nav janvāra un aprīļa ierakstu — ML modelis nebūs "redzējis" šo mēnešu uzvedības modeļus.
Datos ir daudz ekstrēmo vērtību — daži apmeklētāji uzvedas ļoti atšķirīgi no tipiskā apmeklētāja.
Dati ir nelīdzsvaroti — tikai 15.47% apmeklētāju nopirka preci.

Veiktās darbības:
Pārbaudītas trūkstošās vērtības — nav nevienas, dati ir tīri
Weekend un Revenue kodētas no True/False uz 1/0.
Month un VisitorType kodētas ar One-Hot Encoding - no 2 teksta kolonnām izveidotas 11 jaunas 0/1 kolonnas.
26 skaitliskās kolonnas normalizētas ar StandardScaler: vidējais ≈ 0, standartnovirze ≈ 1.
Izveidota jauna iezīme Engagement_Score = ProductRelated_Duration - BounceRates.

Rezultāts:
Sākotnējā datu kopa ar 18 kolonnām apstrādāta un paplašināta līdz 28 kolonnām. Visas kolonnas ir skaitliskas un gatavas mašīnmācīšanās modeļu trenēšanai. Tīrā datu kopa saglabāta failā shoppers_clean.csv.
